```python
class CustomDataset(Dataset):
    def __init__(self, image_paths, labels, transform=None):
        self.image_paths = image_paths
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        image = Image.open(self.image_paths[idx]).convert('RGB')
        if self.transform:
            image = self.transform(image)        
        label = torch.tensor(self.labels[idx], dtype=torch.long)
        return image, label
```

```python
def calculate_accuracy(model, data_loader):
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for inputs, labels in tqdm(data_loader):
            inputs = inputs.to(device)
            labels = labels.to(device)
            outputs = model(inputs)
            
            probs = torch.sigmoid(outputs)
            _, predicted = torch.max(probs, 1)
            
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    
    accuracy = correct / total * 100.0
    return accuracy
```

```python
def get_feature_explanation(pre_processed_image, feature_num, model=model, device=device):
    # Move the input tensor to the same device as the model
    input_tensor = pre_processed_image[None].to(device).requires_grad_()
    with model.explanation_mode():
        out = model(input_tensor)
        indices = torch.tensor([feature_num]).to(device)
        prediction_logit = torch.gather(out, 1, indices.unsqueeze(1)).squeeze(1)
        prediction_logit.backward(inputs=[input_tensor])
    dynamic_linear_weights = input_tensor.grad
    explanation = model.gradient_to_image(input_tensor[0], dynamic_linear_weights[0])
    return explanation

def get_saliency_of_explanation(pre_processed_image,feature_num,model=model,tr_power = 1):
    sal = get_feature_explanation(pre_processed_image,feature_num,model)[:,:,3]
    sal = np.power(sal, tr_power)
    return sal

def preprocessed_to_display_image(img):
    return img[:3,:,:].permute(1,2,0).cpu().detach().numpy()
```

```python
class ROIDataset(Dataset):
    def __init__(self, image_dir, transform=None):
        self.image_dir = image_dir
        self.image_paths = [os.path.join(image_dir, f) for f in os.listdir(image_dir) if f.lower().endswith('.png')]
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        image = Image.open(img_path).convert('RGB')
        if self.transform:
            image = self.transform(image)
        return image, img_path  # return the path for identification if needed
```

```python
def annotation_to_mask(image_np):
    hsv = cv2.cvtColor(image_np, cv2.COLOR_RGB2HSV)
    
    lower_red1 = np.array([0, 70, 50])
    upper_red1 = np.array([10, 255, 255])
    mask1 = cv2.inRange(hsv, lower_red1, upper_red1)
    
    lower_red2 = np.array([170, 70, 50])
    upper_red2 = np.array([180, 255, 255])
    mask2 = cv2.inRange(hsv, lower_red2, upper_red2)
    
    red_mask = cv2.bitwise_or(mask1, mask2)
    
    kernel = np.ones((8, 8), np.uint8)
    closed_mask = cv2.dilate(red_mask, kernel, iterations=2)
    
    
    inverted = cv2.bitwise_not(closed_mask)
    contours, _ = cv2.findContours(closed_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    filled = np.zeros_like(red_mask)  # same dimensions as the mask
    cv2.drawContours(filled, contours, -1, color=255, thickness=cv2.FILLED)

    binary_filled = filled // 255
    return binary_filled
```

```python
def saliency_to_mask(image, threshold):
    binary_mask = (image > threshold).astype(np.uint8)
    return binary_mask
```

```python
import numpy as np

def compute_metrics(mask_A, mask_B):
    # Compute Intersection and Union for IoU.
    intersection = np.logical_and(mask_A, mask_B)
    union = np.logical_or(mask_A, mask_B)
    
    # Intersection over Union (IoU)
    iou = np.sum(intersection) / np.sum(union)
    
    # Intersection over Total Image Area
    total_pixels = mask_A.shape[0] * mask_A.shape[1]
    intersection_over_total = np.sum(intersection) / total_pixels
    
    # Dice Coefficient
    dice = (2. * np.sum(intersection)) / (np.sum(mask_A) + np.sum(mask_B))
    
    # Pixel-level Accuracy
    correct_pixels = np.sum(np.logical_and(mask_A, mask_B)) + np.sum(np.logical_and(~mask_A, ~mask_B))
    accuracy = correct_pixels / total_pixels
    
    metrics = {
        "IoU": iou,
        "Intersection_over_total": intersection_over_total,
        "Dice": dice,
        "Accuracy": accuracy
    }
    
    return metrics
```

```python
IOUs = []
Dices = []
for image_path in annotations:
    annot_image = Image.open('./ROI_true/' + image_path).convert("RGB")
    mask = annotation_to_mask(np.array(annot_image))    
    mask_pil = Image.fromarray((mask.astype(np.uint8) * 255))
    mask_resized = mask_pil.resize((224, 224), resample=Image.NEAREST)
    mask = (np.array(mask_resized) // 255).astype(bool)

    
    image_2 = Image.open('./images_eval/' + image_path).convert("RGB")
    im = model.transform(image_2)
    s1 = get_saliency_of_explanation(im, 0)
    s2 = get_saliency_of_explanation(im, 1)
    plot_im = preprocessed_to_display_image(im)[:, :, 0]
    mask_2 = saliency_to_mask(s1, 0.45)

    metrics = compute_metrics(mask, mask_2)
    IOUs.append(metrics['IoU'])
    Dices.append(metrics['Dice'])

    
    print("Quantitative Metrics:")
    for key, value in metrics.items():
        print(f"{key}: {value:.4f}")
    print ('\n')
```
