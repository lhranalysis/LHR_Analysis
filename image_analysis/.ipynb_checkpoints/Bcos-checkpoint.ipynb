```python
Norm_tensor = torch.tensor(Norm, dtype=torch.float32)
Out_tensor = torch.tensor(Out, dtype=torch.float32)

X_data = torch.cat((Norm_tensor, Out_tensor), dim=0)
Y_data = torch.cat((torch.zeros(Norm_tensor.size(0)), torch.ones(Out_tensor.size(0)))).long()

X_train, X_test, Y_train, Y_test = train_test_split(X_data, Y_data, test_size=0.2, random_state=42)
train_data = torch.utils.data.TensorDataset(X_train, Y_train)
test_data = torch.utils.data.TensorDataset(X_test, Y_test)
train_loader = torch.utils.data.DataLoader(train_data, batch_size=16, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_data, batch_size=16, shuffle=False)
```

```python
class NormedLinear(nn.Linear):
    def forward(self, in_tensor):
        # Normalize the weights
        w = self.weight / self.weight.norm(p=2, dim=1, keepdim=True)
        return F.linear(in_tensor, w, self.bias)

class BcosLinear(nn.Module):
    def __init__(self, in_features, out_features, b=2, max_out=2, scale=None, scale_fact=100):
        super(BcosLinear, self).__init__()
        
        self.linear = NormedLinear(in_features, out_features * max_out, bias=False)
        self.out_features = out_features * max_out
        self.b = b
        self.max_out = max_out
        self.in_features = in_features
        self.scale = scale or (torch.sqrt(torch.tensor(in_features)) / scale_fact)
        self.detach = False

    def forward(self, in_tensor):
        if self.b == 2:
            return self.fwd_2(in_tensor)
        return self.fwd_b(in_tensor)

    def explanation_mode(self, detach=True):
        self.detach = detach

    def fwd_b(self, in_tensor):
        out = self.linear(in_tensor)
        if self.max_out > 1:
            out = out.view(out.size(0), -1, self.max_out)
            out = out.max(dim=2)[0]
        if self.b == 1:
            return out / self.scale
        norm = in_tensor.norm(p=2, dim=1, keepdim=True) + 1e-6
        abs_cos = (out / norm).abs() + 1e-6
        if self.detach:
            abs_cos = abs_cos.detach()
        out = out * abs_cos.pow(self.b - 1)
        return out / self.scale

    def fwd_2(self, in_tensor):
        out = self.linear(in_tensor)
        if self.max_out > 1:
            out = out.view(out.size(0), -1, self.max_out)
            out = out.max(dim=2)[0]
        norm = in_tensor.norm(p=2, dim=1, keepdim=True) + 1e-6
        if self.detach:
            out = (out * out.abs().detach())
            norm = norm.detach()
        else:
            out = (out * out.abs())
        return out / (norm * self.scale)
```

```python
class BcosNetwork(nn.Module):
    def __init__(self, input_size, output_size):
        super(BcosNetwork, self).__init__()
        self.bcos1 = BcosLinear(input_size, 64)
        self.bcos2 = BcosLinear(64, output_size)

    def forward(self, x):
        x = self.bcos1(x)
        x = self.bcos2(x)
        return torch.sigmoid(x)
```

```python
input_size = 1500
output_size = 1
model = BcosNetwork(input_size, output_size)
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
# Train the model and evaluate at each epoch
num_epochs = 1000
Net_acc = [0]
Net_loss = [np.inf]
for epoch in range(num_epochs):
    model.train()
    for inputs, labels in train_loader:
        outputs = model(inputs).squeeze()
        loss = criterion(outputs, labels.float())
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    model.eval()
    all_preds = []
    if epoch%10 == 0:
        with torch.no_grad():
            for inputs, labels in test_loader:
                outputs = model(inputs).squeeze()
                preds = (outputs > 0.5).int()
                all_preds.extend(preds.tolist())
        accuracy = accuracy_score(Y_test.numpy(), all_preds)
        if accuracy>np.max(Net_acc):
            torch.save(model.state_dict(), 'simple_nn_model.pth')
            print (accuracy)
        Net_acc.append(accuracy)
        Net_loss.append(loss.item())

model.load_state_dict(torch.load('./simple_nn_model.pth'))
```

```python

all_preds = []
all_layer_explanations = [[] for _ in model.children() if isinstance(_, BcosLinear)]  # Create a list for each layer

with torch.no_grad():
    for inputs, labels in test_loader:
        outputs = model(inputs).squeeze()
        preds = (outputs > 0.5).int()
        all_preds.extend(preds.tolist())
        x = inputs
        for i, layer in enumerate(model.children()):
            x = layer(x)
            if isinstance(layer, BcosLinear):
                all_layer_explanations[i].append(x.cpu().numpy())

all_layer_explanations = [np.concatenate(expl, axis=0) for expl in all_layer_explanations]
for idx, layer_expl in enumerate(all_layer_explanations):
    print(f"Layer {idx + 1} explanations shape: {layer_expl.shape}")

accuracy = accuracy_score(Y_test.numpy(), all_preds)
print(f'Test Accuracy: {accuracy * 100:.2f}%')
result = np.where(np.array(all_preds) == Y_test.numpy(), 1, 0)
result_repeated = np.repeat(result[:, np.newaxis], 1500, axis=1)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import pandas as pd
```

```python
data = np.vstack((Norm, Out))
true_labels = np.concatenate((np.zeros(Norm.shape[0]), np.ones(Out.shape[0])))
range_K = range(2, 11)  

silhouette_scores = []
best_K = 2
best_score = -1
best_labels = None

for K in range_K:
    kmeans = KMeans(n_clusters=K, n_init=10, random_state=42)
    labels = kmeans.fit_predict(data)
    score = silhouette_score(data, labels)
    silhouette_scores.append(score)
    df = pd.DataFrame({'Cluster': labels, 'True Label': true_labels})
    cluster_composition = df.groupby(['Cluster', 'True Label']).size().unstack(fill_value=0)
    cluster_composition['C1'] = cluster_composition[0] / (cluster_composition[0] + cluster_composition[1])
    cluster_composition['C2'] = cluster_composition[1] / (cluster_composition[0] + cluster_composition[1])
    print(cluster_composition[['C1', 'C2']])
    print ('\n')


    
    if score > best_score:
        best_K = K
        best_score = score
        best_labels = labels
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range_K, silhouette_scores, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Different Values of K')
plt.grid(True)
plt.show()

print(f"The optimal number of clusters is: {best_K}")
```
