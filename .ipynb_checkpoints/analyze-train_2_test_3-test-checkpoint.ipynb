```python
labels = np.vstack([
    np.tile(0, (control.shape[0], 1)),        
    np.tile(1, (treated_mutant.shape[0], 1)),   
    np.tile(2, (mutant.shape[0], 1))])
labels = labels.flatten()
data = np.vstack([control, treated_mutant, mutant])
print(labels.shape) 
print (data.shape)

```

```python
def plot_2d(x,labels = labels):
    control_transform = x[labels == 0]
    treated_mutant_transform = x[labels == 1]
    mutant_transform = x[labels == 2]

    plt.figure(figsize=(8, 6))
    plt.scatter(control_transform[:, 0], control_transform[:, 1],
                label='Control', color='blue', alpha=0.7)
    plt.scatter(treated_mutant_transform[:, 0], treated_mutant_transform[:, 1],
                label='Treated Mutant', color='green', alpha=0.7)
    plt.scatter(mutant_transform[:, 0], mutant_transform[:, 1],
                label='Mutant', color='red', alpha=0.7)

    # Labeling the plot
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title('2D Scatter Plot of Transformed Data')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
class SimpleNN(nn.Module):
    def __init__(self, input_size, output_size):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(input_size, 128)  
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(128, output_size)  
        self.sigmoid = nn.Sigmoid() 
        
    def forward(self, x):
        x1 = self.fc1(x)      
        x = self.relu(x1)    
        x = self.fc2(x)      
        return x1, x        

input_size = 1500
output_size = 3
model = SimpleNN(input_size, output_size)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, labels,test_size=0.2,random_state=42,stratify=labels)
```

```python
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.long)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test, dtype=torch.long)

train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

batch_size = 32
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

```python
num_epochs = 10000
Net_acc = [0.83]
Net_loss = [np.inf]
epochs_since_improvement = 0
patience = 30

for epoch in range(num_epochs):
    model.train()
    for inputs, labels in train_loader:
        _,outputs = model(inputs)
        loss = criterion(outputs, labels)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    if epoch%100 == 0:
        model.eval()
        all_preds = []
        all_targets = []
        with torch.no_grad():
            for inputs, labels in test_loader:
                _,outputs = model(inputs)
                preds = torch.argmax(outputs, dim=1)
                all_preds.extend(preds.cpu().numpy())
                all_targets.extend(labels.cpu().numpy())
        accuracy = accuracy_score(np.array(all_targets), np.array(all_preds))
        print (accuracy)
        if accuracy>np.max(Net_acc):
            epochs_since_improvement = 0
            torch.save(model.state_dict(), 'NN_3_class.pth')
            print ('Model saved')
        else:
            epochs_since_improvement += 1

        Net_acc.append(accuracy)
        Net_loss.append(loss.item())
        
        if epochs_since_improvement >= patience:
            print(f"Early stopping triggered. No improvement in the last {patience} evaluations.")
            break

model.load_state_dict(torch.load('./NN_3_class.pth'))
```

```python
def extract_nn_features(loader, model = model):
    model.eval()
    all_nn_features = []
    all_labels = []
    with torch.no_grad():
        for batch_data, batch_labels in loader:
            features, _ = model(batch_data)
            all_nn_features.append(features.numpy())
            all_labels.append(batch_labels.numpy())
    nn_features_array = np.concatenate(all_nn_features, axis=0)
    labels_array = np.concatenate(all_labels, axis=0)
    return labels_array, nn_features_array
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import pandas as pd
```

```python
np.shape(data),np.shape(labels_list)
```

```python
true_labels =labels_list

range_K = range(2, 11)  
silhouette_scores = []
best_K = 2
best_score = -1
best_labels = None

for K in range_K:
    kmeans = KMeans(n_clusters=K, n_init=10, random_state=42)
    labels = kmeans.fit_predict(data)
    score = silhouette_score(data, labels)
    silhouette_scores.append(score)

    # Create DataFrame to analyze cluster composition
    df = pd.DataFrame({'Cluster': labels, 'True Label': true_labels})
    # Group by cluster and true label, then count occurrences
    cluster_composition = df.groupby(['Cluster', 'True Label']).size().unstack(fill_value=0)

    # Calculate total size of each cluster.
    # Make sure to add counts for labels 0, 1, and 2.
    cluster_composition['Total Size'] = (
        cluster_composition.get(0, 0) + 
        cluster_composition.get(1, 0) + 
        cluster_composition.get(2, 0)
    )

    # Compute the homogeneity percentages for each true label within each cluster.
    # Using .get(key, 0) ensures that if a cluster does not have one of the labels, it defaults to 0.
    cluster_composition['C1'] = cluster_composition.get(0, 0) / cluster_composition['Total Size']
    cluster_composition['C2'] = cluster_composition.get(1, 0) / cluster_composition['Total Size']
    cluster_composition['C3'] = cluster_composition.get(2, 0) / cluster_composition['Total Size']

    # Print cluster size and homogeneity results.
    print(f"Results for K={K}")
    print(cluster_composition[['Total Size', 'C1', 'C2', 'C3']])
    print('\n')

    # Update the best clustering configuration based on silhouette score.
    if score > best_score:
        best_K = K
        best_score = score
        best_labels = labels

print(f"Best number of clusters: {best_K} with a silhouette score of {best_score:.3f}")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range_K, silhouette_scores, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Different Values of K')
plt.grid(True)
plt.show()

print(f"The optimal number of clusters is: {best_K}")
```
