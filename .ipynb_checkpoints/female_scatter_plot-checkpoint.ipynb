```python
files = os.listdir('./female/')
files = [i for i in files if i.endswith('.mat')]
```

```python
label_1 = [i.split(' ')[1] for i in files]
label_1 = [0 if i=='Control' else 1 for i in label_1]

label_2 = [i.split(' ')[5] for i in files]
label_2 = [0 if i=='LHRchap' else 1 for i in label_2]

label = [0 if l1 == 0 else (1 if l2 == 0 else 2) for l1, l2 in zip(label_1, label_2)]
```

```python
female_control = []
female_treated_mutant = []
female_mutant = []

control_files = []
treated_files = []
mutant_files = []

for i,ele in enumerate(files):
    mat = scipy.io.loadmat('./female/'+ele)
    k = list(mat.keys())[-1]
    if label[i]==0:
        control_files.append([ele for k in range(np.shape(mat[k].T)[0])])
        female_control.append(mat[k].T)
    elif label[i] ==1:
        treated_files.append([ele for k in range(np.shape(mat[k].T)[0])])
        female_treated_mutant.append(mat[k].T)
    elif label[i] ==2:
        mutant_files.append([ele for k in range(np.shape(mat[k].T)[0])])
        female_mutant.append(mat[k].T)
```

```python
def plot_stats(x, data, subplot_index, label, color, a = 1,b = 2):
    mean_val = np.mean(data, axis=0)
    std_val = np.std(data, axis=0)    
    plt.subplot(a, b, subplot_index)
    plt.plot(x, mean_val, color=color, label=label, linewidth=2)
    plt.fill_between(x, mean_val - std_val, mean_val + std_val, color=color, alpha=0.3)
    plt.ylim([-0.5, 1.2])
    plt.xlabel('Time',  fontsize=12)
    plt.ylabel(r'$\Delta F / F_0$',  fontsize=12)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.legend(fontsize =12)
```

```python
def remove_outliers(data, threshold=1.5):
    Q1 = np.percentile(data, 25, axis=0)
    Q3 = np.percentile(data, 75, axis=0)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR
    
    mask = np.all((data >= lower_bound) & (data <= upper_bound), axis=1)
    return data[mask], mask
```

```python
def get_clusters(data, true_labels, BK=2):
    range_K = range(2, 7)
    best_K = []
    best_score = []
    best_labels = []
    best_kmeans = []

    for K in range_K:
        kmeans = KMeans(n_clusters=K, n_init=10, random_state=42).fit(data)
        labels = kmeans.labels_
        score = silhouette_score(data, labels)
        print(f"Results for K={K}, Silhouette Score: {score:.3f}")
        
        df = pd.DataFrame({'Cluster': labels, 'True Label': true_labels})
        cluster_composition = df.groupby(['Cluster', 'True Label']).size().unstack(fill_value=0)

        cluster_composition['Total Size'] = (
            cluster_composition.get(0, 0) +  # Het
            cluster_composition.get(1, 0) +  # Rescue
            cluster_composition.get(2, 0))   # Hom

        cluster_composition['C_Het'] = cluster_composition.get(0, 0) / cluster_composition['Total Size']
        cluster_composition['C_Rescue'] = cluster_composition.get(1, 0) / cluster_composition['Total Size']
        cluster_composition['C_Hom'] = cluster_composition.get(2, 0) / cluster_composition['Total Size']

        print(cluster_composition[['Total Size', 'C_Het', 'C_Rescue', 'C_Hom']], "\n")
        
        best_K.append(K)
        best_score.append(score)
        best_labels.append(labels)
        best_kmeans.append(kmeans)

    print([np.shape(i) for i in [best_K, best_score, best_labels, best_kmeans]])
    return best_K[BK], best_score[BK], best_labels[BK], best_kmeans[BK]

def plot_clusters_with_std(data, kmeans,labels):
    centers = kmeans.cluster_centers_
    labels = kmeans.labels_
    n_clusters = centers.shape[0]
    plt.figure(figsize=(10, 8))    
    stds = np.array([np.std(data[labels == i], axis=0) for i in range(n_clusters)])
    
    for i, center in enumerate(centers):
        cluster_data = data[labels == i]
        plt.scatter(cluster_data[:, 0], cluster_data[:, 1], s=30, label=f"Cluster {i}", alpha=0.6)
        
        ellipse = Ellipse(xy=center, width=3*stds[i, 0], height=3*stds[i, 1],edgecolor='black', fc='lightblue', alpha=0.3, lw=2)
        plt.gca().add_patch(ellipse)
        plt.scatter(center[0], center[1], color='red', marker='x', s=100, linewidths=3)
    
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.legend()
    plt.show()

def plotter(data_2d, labels_array):
    x_min, x_max = np.min(data_2d[:, 0]), np.max(data_2d[:, 0])
    bin_width = (x_max - x_min) / 7
    x_intervals = np.linspace(x_min, x_max, 8)  # Ensure exactly 7 bins
    
    colors = ['red', 'green', 'blue']
    fig, ax = plt.subplots(figsize=(10, 6))

    control_transform = data_2d[labels_array == 0]
    treated_mutant_transform = data_2d[labels_array == 1]
    mutant_transform = data_2d[labels_array == 2]

    plt.scatter(control_transform[:, 0], control_transform[:, 1],label='Het', color='blue', alpha=0.8,edgecolors='k')
    plt.scatter(mutant_transform[:, 0], mutant_transform[:, 1],label='Hom', color='red', alpha=0.8,edgecolors='k')
    plt.scatter(treated_mutant_transform[:, 0], treated_mutant_transform[:, 1],label='Hom + LHR-Chap', color='green', alpha=0.8,edgecolors='k')
    
    for x in x_intervals:
        ax.axvline(x, color='gray', linestyle='--', alpha=0.7)
    
    # Adjust y limits dynamically
    y_min, y_max = np.min(data_2d[:, 1]), np.max(data_2d[:, 1])
    ax.set_ylim(y_min - ((y_max - y_min) * 0.2), y_max + (y_max - y_min) * 0.2)
    
    for i in range(len(x_intervals) - 1):
        x_start, x_end = x_intervals[i], x_intervals[i + 1]
        if i == len(x_intervals) - 2:
            mask = (data_2d[:, 0] >= x_start) & (data_2d[:, 0] <= x_end)  # Include the last bin
        else:
            mask = (data_2d[:, 0] >= x_start) & (data_2d[:, 0] < x_end)
        counts = np.bincount(labels_array[mask], minlength=3)  # Count occurrences of labels 0,1,2
        text_x = (x_start + x_end) / 2  # Midpoint of the interval
        text_y = y_max + (y_max - y_min) * 0.1  # Position much above the plot
        ax.text(text_x, text_y, f"{counts[0]},{counts[1]},{counts[2]}",
                ha='center', fontsize=10, bbox=dict(facecolor='white', alpha=0.5))
    
    ax.set_xlabel("Feature 1")
    ax.set_ylabel("Feature 2")
    # ax.set_title("Scatter Plot")
    plt.legend(loc='lower left')

    plt.show()
```

```python
def plotter(data_2d, labels_array):
    import numpy as np
    import matplotlib.pyplot as plt
    from matplotlib.offsetbox import TextArea, HPacker, AnnotationBbox

    x_min, x_max = np.min(data_2d[:, 0]), np.max(data_2d[:, 0])
    x_intervals = np.linspace(x_min, x_max, 8)  # 7 bins

    fig, ax = plt.subplots(figsize=(10, 6))

    control_transform = data_2d[labels_array == 0]
    treated_mutant_transform = data_2d[labels_array == 1]
    mutant_transform = data_2d[labels_array == 2]

    ax.scatter(control_transform[:, 0], control_transform[:, 1],
               label='Het', color='blue', alpha=0.8, edgecolors='k')
    ax.scatter(mutant_transform[:, 0], mutant_transform[:, 1],
               label='Hom', color='red', alpha=0.8, edgecolors='k')
    ax.scatter(treated_mutant_transform[:, 0], treated_mutant_transform[:, 1],
               label='Hom + LHR-Chap', color='green', alpha=0.8, edgecolors='k')

    # Draw vertical bin lines.
    for x in x_intervals:
        ax.axvline(x, color='gray', linestyle='--', alpha=0.7)

    y_min, y_max = np.min(data_2d[:, 1]), np.max(data_2d[:, 1])
    ax.set_ylim(y_min - 0.2 * (y_max - y_min), y_max + 0.2 * (y_max - y_min))

    # Function to format normalized values.
    def format_norm(val):
        s = f"{val:.2f}"
        return s[1:] if s.startswith("0") else s

    for i in range(len(x_intervals) - 1):
        x_start, x_end = x_intervals[i], x_intervals[i + 1]
        if i == len(x_intervals) - 2:
            mask = (data_2d[:, 0] >= x_start) & (data_2d[:, 0] <= x_end)
        else:
            mask = (data_2d[:, 0] >= x_start) & (data_2d[:, 0] < x_end)
        counts = np.bincount(labels_array[mask], minlength=3)
        total = counts.sum()
        if total > 0:
            norm0 = counts[0] / total
            norm1 = counts[1] / total
            norm2 = counts[2] / total
        else:
            norm0 = norm1 = norm2 = 0

        # Original code swapped the order for norm1 and norm2.
        blue_text = format_norm(norm0)
        red_text = format_norm(norm2)
        green_text = format_norm(norm1)

        center = (x_start + x_end) / 2
        text_y = y_max + 0.1 * (y_max - y_min)

        # Create individual text areas for each fraction.
        blue_area = TextArea(blue_text, textprops=dict(color='blue', fontsize=10))
        red_area = TextArea(red_text, textprops=dict(color='red', fontsize=10))
        green_area = TextArea(green_text, textprops=dict(color='green', fontsize=10))
        
        # Create comma TextAreas to be placed between the values.
        comma1 = TextArea(", ", textprops=dict(color='black', fontsize=10))
        comma2 = TextArea(", ", textprops=dict(color='black', fontsize=10))

        # Pack the three fraction values with commas into one horizontal box.
        packed_text = HPacker(children=[blue_area, comma1, red_area, comma2, green_area],
                              align="center", pad=0, sep=0)

        # Create a single annotation box for the packed text.
        ab = AnnotationBbox(packed_text, (center, text_y),
                            xycoords='data',
                            box_alignment=(0.5, 0),
                            bboxprops=dict(facecolor='white', alpha=0.5))
        ax.add_artist(ab)

    ax.set_xlabel("Feature 1", fontsize=12)
    ax.set_ylabel("Feature 2", fontsize=12)
    ax.tick_params(labelsize=12)
    ax.legend(loc='lower left', fontsize=12)

    plt.show()
```

```python
class SimpleNN(nn.Module):
    def __init__(self, input_size, output_size):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(input_size, 2)  
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(2, output_size)  
        self.sigmoid = nn.Sigmoid() 
        
    def forward(self, x):
        x1 = self.fc1(x)      
        x = self.relu(x1)    
        x = self.fc2(x)      
        return x1, x        

class SimpleNN_2(nn.Module):
    def __init__(self, input_size, output_size):
        super(SimpleNN_2, self).__init__()
        self.fc1 = nn.Linear(input_size, 128)  
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(128, output_size)  
        self.sigmoid = nn.Sigmoid() 
        
    def forward(self, x):
        x1 = self.fc1(x)      
        x = self.relu(x1)    
        x = self.fc2(x)      
        return x1, x        
```

```python
def prep_data(data,labels):
    X_train, X_test, y_train, y_test = train_test_split(data, labels,test_size=0.2,random_state=42,stratify=labels)

    X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
    y_train_tensor = torch.tensor(y_train, dtype=torch.long)
    X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
    y_test_tensor = torch.tensor(y_test, dtype=torch.long)

    train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
    test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

    batch_size = 32
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
    return train_loader, test_loader
```

```python
def test_all(data = data_total, label = label_total):
    X_tot = torch.tensor(data, dtype=torch.float32)
    y_tot = torch.tensor(label, dtype=torch.long)

    tot_dataset = TensorDataset(X_tot, y_tot)
    tot_loader = DataLoader(tot_dataset, batch_size=32, shuffle=False)
    return tot_loader
```

```python
def extract_nn_features(loader, model):
    model.eval()
    all_nn_features = []
    all_labels = []
    with torch.no_grad():
        for batch_data, batch_labels in loader:
            features, _ = model(batch_data)
            all_nn_features.append(features.numpy())
            all_labels.append(batch_labels.numpy())
    nn_features_array = np.concatenate(all_nn_features, axis=0)
    labels_array = np.concatenate(all_labels, axis=0)
    return labels_array, nn_features_array
```

```python
def model_train(model,train_loader,test_loader,name = 3):
    num_epochs = 10000
    Net_acc = [0.0]
    Net_loss = [np.inf]
    epochs_since_improvement = 0
    patience = 30
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    
    for epoch in range(num_epochs):
        model.train()
        for inputs, labels in train_loader:
            _,outputs = model(inputs)
            loss = criterion(outputs, labels)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
    
        if epoch%50 == 0:
            model.eval()
            all_preds = []
            all_targets = []
            with torch.no_grad():
                for inputs, labels in test_loader:
                    _,outputs = model(inputs)
                    preds = torch.argmax(outputs, dim=1)
                    all_preds.extend(preds.cpu().numpy())
                    all_targets.extend(labels.cpu().numpy())
            accuracy = accuracy_score(np.array(all_targets), np.array(all_preds))
            print (accuracy)
            if accuracy>np.max(Net_acc):
                epochs_since_improvement = 0
                torch.save(model.state_dict(), './models/save_model'+str(name)+'.pth')
                print ('Model saved')
            else:
                epochs_since_improvement += 1
    
            Net_acc.append(accuracy)
            Net_loss.append(loss.item())
            
            if epochs_since_improvement >= patience:
                print(f"Early stopping triggered. No improvement in the last {patience} evaluations.")
                break    
    model.load_state_dict(torch.load('./models/save_model'+str(name)+'.pth'))
    return model
```

```python
def plot_clusters(data, true_labels, BK=2):
    range_K = range(2, 8)
    best_K = []
    best_score = []
    best_labels = []
    best_kmeans = []
    
    fig, axs = plt.subplots(2, 2, figsize=(10, 10))
    axs = axs.flatten()
    
    for idx, K in enumerate([2, 3, 4, 5]):
        kmeans = KMeans(n_clusters=K, n_init=10, random_state=42).fit(data)
        labels = kmeans.labels_
        score = silhouette_score(data, labels)
        print(f"Results for K={K}, Silhouette Score: {score:.3f}")
        
        # Plot scatter using the first two features on the corresponding subplot
        ax = axs[idx]
        scatter = ax.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', edgecolor='k',alpha = 0.7)
        ax.set_title(f"KMeans Clustering (K={K})", fontsize=14)
        ax.set_xlabel("Feature 1", fontsize=12)
        ax.set_ylabel("Feature 2", fontsize=12)
        ax.tick_params(axis='both', labelsize=12)
        ax.grid(True, linestyle='--', alpha=0.5)
        
        # Create a legend based on unique clusters
        handles, _ = scatter.legend_elements()
        ax.legend(handles, [f"Cluster {i}" for i in range(K)], fontsize=12, title="Clusters")
        
        # Build a DataFrame to compare k-means clusters with the true labels
        df = pd.DataFrame({'Cluster': labels, 'True Label': true_labels})
        cluster_composition = df.groupby(['Cluster', 'True Label']).size().unstack(fill_value=0)
        
        # For 3 true classes (0, 1, and 2)
        cluster_composition['Total Size'] = (
            cluster_composition.get(0, 0) +
            cluster_composition.get(1, 0) +
            cluster_composition.get(2, 0)
        )
        cluster_composition['C0'] = cluster_composition.get(0, 0) / cluster_composition['Total Size']
        cluster_composition['C1'] = cluster_composition.get(1, 0) / cluster_composition['Total Size']
        cluster_composition['C2'] = cluster_composition.get(2, 0) / cluster_composition['Total Size']
        print(cluster_composition[['Total Size', 'C0', 'C1', 'C2']], "\n")
        
        best_K.append(K)
        best_score.append(score)
        best_labels.append(labels)
        best_kmeans.append(kmeans)
    
    plt.tight_layout()
    plt.show()
    
    return best_K[BK], best_score[BK], best_labels[BK], best_kmeans[BK]
```
