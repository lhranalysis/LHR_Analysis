```python
Norm_tensor = torch.tensor(Norm, dtype=torch.float32)
Out_tensor = torch.tensor(Out, dtype=torch.float32)

# Create X_data and Y_data
X_data = torch.cat((Norm_tensor, Out_tensor), dim=0)
Y_data = torch.cat((torch.zeros(Norm_tensor.size(0)), torch.ones(Out_tensor.size(0)))).long()

# Train-test split
X_train, X_test, Y_train, Y_test = train_test_split(X_data, Y_data, test_size=0.2, random_state=42)

# Convert to PyTorch DataLoader for batching
train_data = torch.utils.data.TensorDataset(X_train, Y_train)
test_data = torch.utils.data.TensorDataset(X_test, Y_test)

train_loader = torch.utils.data.DataLoader(train_data, batch_size=16, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_data, batch_size=16, shuffle=False)
```

```python
# Define a simple neural network model
class SimpleNN(nn.Module):
    def __init__(self, input_size, output_size):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(input_size, 64)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(64, output_size)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.sigmoid(x)
        return x
```

```python
# Initialize model, loss function, and optimizer
input_size = 1500
output_size = 1
model = SimpleNN(input_size, output_size)
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
# Train the model and evaluate at each epoch
num_epochs = 10000
Net_acc = [0.86]
Net_loss = [np.inf]
for epoch in range(num_epochs):
    model.train()
    for inputs, labels in train_loader:
        outputs = model(inputs).squeeze()
        loss = criterion(outputs, labels.float())
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    model.eval()
    all_preds = []
    if epoch%100 == 0:
        with torch.no_grad():
            for inputs, labels in test_loader:
                outputs = model(inputs).squeeze()
                preds = (outputs > 0.5).int()
                all_preds.extend(preds.tolist())
        accuracy = accuracy_score(Y_test.numpy(), all_preds)
        print (accuracy)
        if accuracy>np.max(Net_acc):
            torch.save(model.state_dict(), '1simple_nn_model.pth')
            print (epoch,accuracy)
        Net_acc.append(accuracy)
        Net_loss.append(loss.item())

model.load_state_dict(torch.load('./1simple_nn_model.pth'))
```

```python
def get_feature(inp):
    with torch.no_grad():
        sample_input = inp
        features = model.fc1(sample_input)  
    return(features)  # Should print torch.Size([1, 64])
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import pandas as pd
```

```python
# Stack the data
data = np.vstack((Norm, Out))
true_labels = np.concatenate((np.zeros(Norm.shape[0]), np.ones(Out.shape[0])))

range_K = range(2, 11)  
silhouette_scores = []
best_K = 2
best_score = -1
best_labels = None

for K in range_K:
    kmeans = KMeans(n_clusters=K, n_init=10, random_state=42)
    labels = kmeans.fit_predict(data)
    score = silhouette_score(data, labels)
    silhouette_scores.append(score)

    # Create DataFrame to analyze cluster composition
    df = pd.DataFrame({'Cluster': labels, 'True Label': true_labels})
    cluster_composition = df.groupby(['Cluster', 'True Label']).size().unstack(fill_value=0)

    # Total size of each cluster
    cluster_composition['Total Size'] = cluster_composition[0] + cluster_composition[1]

    # Homogeneity: percentage of each true class in a cluster
    cluster_composition['C1'] = cluster_composition[0] / cluster_composition['Total Size']
    cluster_composition['C2'] = cluster_composition[1] / cluster_composition['Total Size']

    # Print cluster size and homogeneity
    print(f"Results for K={K}")
    print(cluster_composition[['Total Size', 'C1', 'C2']])
    print('\n')

    # Update best clustering configuration
    if score > best_score:
        best_K = K
        best_score = score
        best_labels = labels
```
